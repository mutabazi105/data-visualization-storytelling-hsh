## 📊 Schritt 3: Erstelle deine Visualisierung

````python
# Erstelle hier deinen Plot:
fig, ax = plt.subplots(figsize=(10, 6))

# Beispiel: Liniendiagramm
# ax.plot(df['x'], df['y'], color='teal', linewidth=2)

# Beschriftungen:
ax.set_title('Dein Titel hier', fontsize=16)
ax.set_xlabel('X-Achse Beschriftung', fontsize=12)
ax.set_ylabel('Y-Achse Beschriftung', fontsize=12)

# Gitter:
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
