## Teil C – Linear vs. logarithmisch

Vergleiche die Verteilung von `revenue` einmal linear und einmal logarithmisch.
Notiere anschließend, welche Darstellung wann sinnvoll ist.


```python
# Starter-Code Teil C
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(retail['revenue'], bins=12)
axes[0].set_title('Revenue – lineare Skala')

axes[1].hist(retail['revenue'], bins=12)
axes[1].set_xscale('log')
axes[1].set_title('Revenue – logarithmische Skala')

plt.tight_layout()
plt.show()
```
