## Abschnitt 9: Linear vs. logarithmisch

Nutze hier ein Beispiel mit `revenue`, um zu zeigen:
- wie sich Verteilungen je nach Skala unterschiedlich lesen
- wann eine Log-Skala hilfreich sein kann
- warum sie immer erklärt werden sollte


```python
# Vergleich linear vs. log
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(retail['revenue'], bins=12)
axes[0].set_title('Revenue – lineare Skala')

axes[1].hist(retail['revenue'], bins=12)
axes[1].set_xscale('log')
axes[1].set_title('Revenue – logarithmische Skala')

plt.tight_layout()
plt.show()
```
