# Intensity Chart Visualization

This notebook creates a visualization showing intensity trends over time periods 22-32, with multiple trend lines (A, B, C, D, Other) and a growth projection curve.

## Chart Components:
- **Line A**: Stable trend around 25-30%
- **Line B**: Declining trend from ~70% to ~40%
- **Line C**: Continuation of decline from ~40% to near 0%
- **Line D**: High volatility around 90-95%
- **Other**: Scattered data points
- **Growth Curve**: Exponential growth projection (green)

## Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import FancyBboxPatch

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

## Define Synthetic Data

Creating data patterns that match the visual requirements for each trend line.

```python
# Line A: Stable around 25-30% (periods 22-24)
line_a_x = np.array([22, 23, 24])
line_a_y = np.array([28, 26, 27])

# Line B: Declining from ~70% to ~40% (periods 23-25)
line_b_x = np.array([23, 24, 25])
line_b_y = np.array([70, 55, 40])

# Line C: Continuing decline from ~40% to near 0% (periods 25-26)
line_c_x = np.array([25, 26])
line_c_y = np.array([40, 5])

# Line D: High volatility around 90-95% (periods 27-28)
line_d_x = np.array([27, 27.3, 27.6, 28])
line_d_y = np.array([92, 95, 90, 93])

# Other: Scattered points (periods 29-30)
other_x = np.array([29, 29.3, 29.7, 30])
other_y = np.array([25, 32, 20, 28])

# Growth curve: Exponential growth from ~50% to ~100% (periods 31-32)
growth_x = np.linspace(31, 32, 50)
# Create smooth exponential growth curve
growth_y = 50 + 50 * (1 - np.exp(-5 * (growth_x - 31)))
```

## Create the Visualization

Building the complete chart with all components, styling, and annotations.

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plot all trend lines in gray
gray_color = '#808080'
line_width = 2

# Plot Line A
ax.plot(line_a_x, line_a_y, color=gray_color, linewidth=line_width, 
        marker='o', markersize=6, label='A')

# Plot Line B
ax.plot(line_b_x, line_b_y, color=gray_color, linewidth=line_width, 
        marker='o', markersize=6, label='B')

# Plot Line C
ax.plot(line_c_x, line_c_y, color=gray_color, linewidth=line_width, 
        marker='o', markersize=6, label='C')

# Plot Line D
ax.plot(line_d_x, line_d_y, color=gray_color, linewidth=line_width, 
        marker='o', markersize=6, label='D')

# Plot Other as scattered points
ax.scatter(other_x, other_y, color=gray_color, s=80, 
           marker='o', label='Other', zorder=5)

# Plot Growth curve in green
green_color = '#2ecc71'
ax.plot(growth_x, growth_y, color=green_color, linewidth=3, 
        label='Growth Projection', zorder=10)

# Add annotation for growth curve
annotation_text = 'Strong stable growth\nexpected to continue'
ax.annotate(annotation_text, 
            xy=(31.5, 75), 
            xytext=(30.5, 60),
            fontsize=11,
            color=green_color,
            weight='bold',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='white', 
                     edgecolor=green_color, linewidth=2),
            arrowprops=dict(arrowstyle='->', color=green_color, lw=2))

# Customize axes
ax.set_xlabel('Time Period', fontsize=12, weight='bold')
ax.set_ylabel('Intensity', fontsize=12, weight='bold')
ax.set_title('Intensity Trends Over Time', fontsize=16, weight='bold', pad=20)

# Set axis limits and ticks
ax.set_xlim(21.5, 32.5)
ax.set_ylim(0, 100)
ax.set_xticks(range(22, 33))
ax.set_yticks(range(0, 101, 10))

# Format y-axis as percentages
ax.set_yticklabels([f'{y}%' for y in range(0, 101, 10)])

# Add grid
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# Add legend
ax.legend(loc='upper left', fontsize=10, framealpha=0.9)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

## Summary

This visualization successfully demonstrates:

1. **Multiple trend lines** (A, B, C, D) showing different patterns:
   - Stability (Line A)
   - Decline (Lines B and C)
   - High volatility (Line D)
   - Scattered data (Other)

2. **Growth projection** shown as a smooth green curve with exponential growth characteristics

3. **Clear annotation** highlighting the expected growth trend

4. **Professional styling** with:
   - Percentage-formatted y-axis (0% to 100%)
   - Time period x-axis (22 to 32)
   - Grid for easy reading
   - Legend for line identification
   - Appropriate colors (gray for historical data, green for projection)

The chart can be easily customized by modifying the data arrays or styling parameters in the code cells above.
