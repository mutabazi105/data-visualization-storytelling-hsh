```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
# Set style
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
```

```python


start = datetime.date(2026, 4, 22)
end   = datetime.date(2026, 5, 7)
current = start
tag_datum = []

i=0
while current <= end:
    tag_datum.append(current.strftime("%d.%m"))  # formats as "22.04", "23.04", etc.
    current += datetime.timedelta(days=1)         # move to next day
```

```python

temp_list_tagsueber= [15,15,14,14,15,14,12,13,13,14,6,10,13,18,16,19]

temp_arr_tagsueber=np.array(temp_list_tagsueber) 

print(len(temp_arr_tagsueber))

temp_list_nachts= [1,2,5,5,6,4,4,6,6,5,4,4,6,5,4,4]

temp_arr_nachts=np.array(temp_list_nachts)


avg_tag=temp_arr_tagsueber.mean()
# mean() Funktion von Python 3.4+; alternativ: sum(temp_list_tagsueber) / len(temp_list_tagsueber)

print(avg_tag)

max_tag=temp_arr_tagsueber.max()

max_tag_index=np.where(temp_arr_tagsueber==max_tag)[0][0]
# find the index of defined value, np.argmax() just for max value's index, 2D array, so [0][0]

min_tag=min(temp_arr_tagsueber)

min_tag_index=np.where(temp_arr_tagsueber==min_tag)[0][0]


avg_nacht=temp_arr_nachts.mean()

max_nacht=temp_arr_nachts.max()

max_nacht_index=np.argmax(temp_arr_nachts)
# argmax（）find the index of max value

min_nacht=temp_arr_nachts.min()

min_nacht_index=np.argmin(temp_arr_nachts) 



fig=plt.figure()

ax_tag=fig.add_axes([0.8,2.2,1.3,1])
#[l, b, w, h]

ax_tag.plot(tag_datum,temp_arr_tagsueber,marker='o',color='g')

ax_tag.set_title(r"22.April bis 7. Mai 2026 - Tagestemperaturen in$^\circ C$")

ax_tag.set_ylabel(r"$^\circC$",loc='top');

ax_tag.set_ylim(0,21)

ax_tag.set_xlabel("Datum",fontsize=12);

ax_tag.grid(True)

ax_tag.axhline(avg_tag,color='r',linestyle='--',label='Durchschnitt')

ax_tag.scatter(tag_datum[max_tag_index],max_tag,color='red',label='Max Tag',zorder=5,s=140)

ax_tag.text(max_tag_index,max_tag+0.5,f'Max:{max_tag}°C',ha='center',va='bottom',color='red')
# ha: horizontal alignment, va: vertical alignment

ax_tag.scatter(tag_datum[min_tag_index],min_tag,color='red',
label='Min Tag',
zorder=4,s=140)

ax_tag.text(min_tag_index,
min_tag-0.5,
f'Min:
{min_tag}°C',
ha='center',
va='top',
color='red')


# add mean value to ticks

current_ticks_tag
=
ax_tag.get_yticks()

add_mean_ticks_tag
=
np.append(current_ticks_tag,
avg_tag)

print(f'current_ticks_tag:
{current_ticks_tag}') 

print(f'add_mean_ticks_tag:
{add_mean_ticks_tag}') 

ax_tag.set_yticks(add_mean_ticks_tag)

"""""where(add_mean_ticks_tag == avg_tag)，compare the value in add_mean_ticks_tag with avg_tag,

 return the index of the value which is equal to avg_tag; [[False, False, False, False, False, False, False, False, False, False, True]], 

 and we need the first element of the tuple, which is an array, and we need the first element of that array, which is the index of avg_tag in add_mean_ticks_tag"""

mean_y_tag_index
=
np.where(add_mean_ticks_tag
==
avg_tag)[0][0]

print(add_mean_ticks_tag[1])
# output: 10

# print(f"mean_y_tag_index: {mean_y_tag_index}") # output: 10

mean_y_tag_index
=
np.where(add_mean_ticks_tag
==
avg_tag)[0][0]

ax_tag.yaxis.get_ticklabels()[mean_y_tag_index].set_color('red')
# change the color of mean value in ticks


ax_nacht
=
fig.add_axes([0.8,
1,1.3,
1])
# 

ax_nacht.plot(tag_datum,
temp_arr_nachts,marker='o',
color='b')

ax_nacht.set_title(r"22.
 April bis 7. Mai 2026 - Nachtestemperaturen in
$^\circ C$
 ")

ax_nacht.grid(True)

ax_nacht.set_xlabel("Datum",fontsize=12);

ax_nacht.set_ylabel(r"$^\circ
 C$",loc
='top');

ax_nacht.set_ylim(0,10)

ax_nacht.axhline(avg_nacht,
color='r',
linestyle='--',
label='Durchschnitt')

ax_nacht.scatter(tag_datum[max_nacht_index],
max_nacht,
color='red',  zorder=5,s=140)

ax_nacht.text(max_nacht_index,
max_nacht+0.5,
f'Max:
{max_nacht}°C',
ha='center',
va='bottom',
color='red')

ax_nacht.scatter(tag_datum[min_nacht_index],
min_nacht,
color='red',  zorder=5,s=140)

ax_nacht.text(min_nacht_index,
min_nacht-0.5,
f'Min:
{min_nacht}°C',
ha='center',
va='top',
color='red')


# add mean value to ticks

current_ticks_nachte
=
ax_nacht.get_yticks()

add_mean_ticks_nacht
=
np.append(current_ticks_nachte,
avg_nacht)

ax_nacht.set_yticks(add_mean_ticks_nacht)

mean_y_nacht_index
=
np.where(add_mean_ticks_nacht
==
avg_nacht)[0][0]

ax_nacht.yaxis.get_ticklabels()[mean_y_nacht_index].set_color('red')
# change the color of mean value in ticks
```
