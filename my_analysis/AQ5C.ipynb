```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simple temperature data for one week
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Hannover': [12, 14, 16, 15, 13, 11, 10],
    'Hamburg': [11, 13, 15, 14, 12, 10, 9]
}

df = pd.DataFrame(data)
print("Temperature Data:")
print(df)
```

```python
# Calculate average temperatures
hannover_avg = df['Hannover'].mean()
hamburg_avg = df['Hamburg'].mean()

print(f"Hannover average: {hannover_avg:.1f}°C")
print(f"Hamburg average: {hamburg_avg:.1f}°C")
print(f"Difference: {hannover_avg - hamburg_avg:.1f}°C")
```

```python
# Find biggest daily difference
df['Diff'] = abs(df['Hannover'] - df['Hamburg'])
max_idx = df['Diff'].idxmax()
max_day = df.loc[max_idx, 'Day']
max_diff = df.loc[max_idx, 'Diff']

print(f"Biggest difference: {max_diff}°C on {max_day}")
```

```python
# Calculate fluctuation (how much temperature changes)
han_range = df['Hannover'].max() - df['Hannover'].min()
ham_range = df['Hamburg'].max() - df['Hamburg'].min()

print(f"Hannover range: {han_range}°C")
print(f"Hamburg range: {ham_range}°C")
```

```python
# Save the data to a CSV file
csv_filename = 'temperature_data.csv'
df.to_csv(csv_filename, index=False)

print(f"✅ CSV file saved as: {csv_filename}")
print(f"📁 Location: {csv_filename}")
print("\n📄 File contents preview:")
print(df)
```

```python
# Verify by reading the CSV file back
df_loaded = pd.read_csv(csv_filename)

print("✅ CSV file loaded successfully!")
print("\n📖 Data from CSV file:")
print(df_loaded)
print(f"\n📊 Shape: {df_loaded.shape[0]} rows × {df_loaded.shape[1]} columns")
```

```python
# Create a simple, clean chart
plt.figure(figsize=(10, 6))

# Plot lines for each city
plt.plot(df['Day'], df['Hannover'], marker='o', linewidth=2, 
         label='Hannover', color='red', markersize=8)
plt.plot(df['Day'], df['Hamburg'], marker='s', linewidth=2, 
         label='Hamburg', color='blue', markersize=8)

# Labels and title
plt.xlabel('Day', fontsize=12, fontweight='bold')
plt.ylabel('Temperature (°C)', fontsize=12, fontweight='bold')
plt.title('Temperature Comparison: Hannover vs Hamburg', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n📊 Summary:")
print(f"• Hannover is {hannover_avg - hamburg_avg:.1f}°C warmer on average")
print(f"• Both cities follow the same pattern")
print(f"• Hannover: {df['Hannover'].min()}°C to {df['Hannover'].max()}°C")
print(f"• Hamburg: {df['Hamburg'].min()}°C to {df['Hamburg'].max()}°C")
```
