```python
import pandas as pd
import numpy as np
from typing import Any

# ============================================
# STEP 1: Define your initial messe_data
# ============================================

messe_data = {
    # Your existing data here (Oleh, Blerina, Bernard, etc.)
    "Oleh":{
        "attended_messe": False,
        "spoke_to_people": False,
        "positions_seen": 0,
        "companies_seen": ["Not so much"],
        "career_fields_seen": ["Data Analysis", "AI"],
        "notes": "Did not attend but interested in Data Analyzer field"
    },
    "Blerina": {
        "attended_messe": False,
        "spoke_to_people": False,
        "positions_seen": 0,
        "companies_seen": [],
        "career_fields_seen": ["Data Analyzer"],
        "notes": "Did not attend but interested in Data Analyzer field"
    },
    # ... add all your other students (Menjung, Hani, Bernard, Anton, etc.)
    "Bernard": {
        "attended_messe": True,
        "spoke_to_people": True,
        "positions_seen": None,
        "companies_seen": ["Amazon", "Microsoft", "Schwarz Digits","Coreflux"],
        "career_fields_seen": ["FullStack"],
        "notes": "Visited companies listed"
    },
    "Anton": {
        "attended_messe": True,
        "spoke_to_people": True,
        "positions_seen": None,
        "companies_seen": ["Samsung", "Bundeswehr", "BMW"],
        "career_fields_seen": ["FullStack"],
        "notes": "Visited companies listed"
    }
}

# ============================================
# STEP 2: Load CSV data
# ============================================

df_csv = pd.read_csv('./Sheet1.csv', delimiter=';', encoding='latin-1')

print("CSV Columns:", df_csv.columns.tolist())
print("\nFirst few rows from CSV:")
print(df_csv.head())

# ============================================
# STEP 3: Function to add CSV data to messe_data
# ============================================

def append_csv_to_messe_data(df_csv, messe_data):
    """
    Adds all students from CSV to messe_data
    """
    count_added = 0
    
    for idx, row in df_csv.iterrows():
        name = row.get('Name')
        
        # Skip if name is missing or already in messe_data
        if pd.isna(name) or name in messe_data:
            continue
        
        # Extract values from CSV (adjust column names as needed)
        attended = row.get('Wart ihr auf der Messe?') == 'Ja'
        spoke = row.get('Mit Leuten gesprochen?') == 'Ja'
        
        # Handle positions_seen (convert to int if possible)
        positions = row.get('Wieviele Leute gesehen?')
        if pd.isna(positions):
            positions = 0
        else:
            try:
                positions = int(positions)
            except:
                positions = 0
        
        # Add to messe_data
        messe_data[name] = {
            "attended_messe": attended,
            "spoke_to_people": spoke,
            "positions_seen": positions,
            "companies_seen": [],  # Will be filled from CSV if available
            "career_fields_seen": [],  # Will be filled from CSV if available
            "notes": "Added from CSV"
        }
        count_added += 1
    
    print(f"✅ Added {count_added} students from CSV")
    return messe_data

# Run the function
messe_data = append_csv_to_messe_data(df_csv, messe_data)

# ============================================
# STEP 4: Convert to DataFrame with safe counting
# ============================================

def safe_len(x):
    """Safely get length of a list, return 0 for non-lists or None"""
    if isinstance(x, list):
        return len(x)
    elif pd.isna(x):
        return 0
    else:
        return 0

# Convert dictionary to DataFrame
df = pd.DataFrame.from_dict(messe_data, orient='index')
df.index.name = 'Student'
df = df.reset_index()

# Add derived columns safely
df['num_companies'] = df['companies_seen'].apply(safe_len)
df['num_career_fields'] = df['career_fields_seen'].apply(safe_len)

# Clean up positions_seen (convert None to 0)
df['positions_seen'] = df['positions_seen'].fillna(0).astype(int)

# ============================================
# STEP 5: Display the results
# ============================================

print("\n" + "="*60)
print("Hannover Messe Survey Data Overview")
print("="*60)
print(df[['Student', 'attended_messe', 'spoke_to_people', 'positions_seen', 
          'num_companies', 'num_career_fields']].head(10))

print("\n" + "="*60)
print(f"Total students in dataset: {len(df)}")
print("="*60)

# ============================================
# STEP 6: Basic analysis
# ============================================

print("\n📊 Quick Analysis:")
print(f"- Students who attended the Messe: {df['attended_messe'].sum()}")
print(f"- Students who spoke to people: {df['spoke_to_people'].sum()}")
print(f"- Total companies seen: {df['num_companies'].sum()}")
print(f"- Average companies per student: {df['num_companies'].mean():.2f}")
```
