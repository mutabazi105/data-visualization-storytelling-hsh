# My Age Versus Intensity of Relationship

Reproduction of the provided chart style (gray background, 0-1 y-axis, colored relationship segments, bottom legend).

![](intensity_chart_reference.png)

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Approximated data points from the screenshot
a_x = np.array([22.25, 22.55, 22.85, 23.10, 23.35])
a_y = np.array([0.28, 0.31, 0.29, 0.30, 0.26])

b_x = np.array([24.05, 24.35, 24.50, 24.75, 25.05, 25.25, 25.50])
b_y = np.array([0.64, 0.66, 0.68, 0.55, 0.45, 0.37, 0.10])

c_x = np.array([24.80, 25.05, 25.25, 25.50])
c_y = np.array([0.40, 0.36, 0.24, 0.10])

d_x = np.array([27.25, 27.55, 27.75, 28.00, 28.25, 28.50])
d_y = np.array([0.83, 0.95, 0.68, 0.92, 0.88, 0.56])

cur_x = np.array([30.25, 30.60, 30.95, 31.25, 31.60, 31.90, 32.15, 32.45])
cur_y = np.array([0.52, 0.55, 0.57, 0.60, 0.65, 0.70, 0.80, 0.87])

other_x = np.array([29.00, 29.25, 29.45, 29.65, 29.75])
other_y = np.array([0.22, 0.26, 0.12, 0.36, 0.37])
```

```python
# Figure and style
fig, ax = plt.subplots(figsize=(8.6, 4.8), dpi=140)

bg = '#b3b3b3'
grid = '#d2d2d2'
ax.set_facecolor(bg)
fig.patch.set_facecolor('white')

# Relationship lines
ax.plot(a_x, a_y, lw=3, color='#e67e22', label='Relationship A')
ax.plot(b_x, b_y, lw=3.2, color='#7030a0', label='Relationship B')
ax.plot(c_x, c_y, lw=3.2, color='#f1c40f', label='Relationship C')
ax.plot(d_x, d_y, lw=3.2, color='#5dade2', label='Relationship D')
ax.plot(cur_x, cur_y, lw=3.2, color='#6ab04c', label='Current')

# Other points
ax.scatter(other_x, other_y, s=34, color='#1f3b73', edgecolors='#2f4f8f', linewidths=0.6, label='Other', zorder=5)

# Axes
ax.set_xlim(22, 32.7)
ax.set_ylim(0, 1.0)
ax.set_xticks(np.arange(22, 33, 1))
yticks = np.arange(0, 1.01, 0.1)
ax.set_yticks(yticks)
ax.set_yticklabels([f'{t:g}' for t in yticks])

# Grid and spines
ax.grid(axis='y', color=grid, linewidth=1.0, alpha=0.9)
ax.grid(axis='x', visible=False)
for sp in ax.spines.values():
    sp.set_color('#b7b7b7')

ax.tick_params(colors='#2f2f2f', labelsize=10)
ax.set_title('My Age Versus Intensity of Relationship', fontsize=18, color='#3e3e3e', pad=16)

# Legend at bottom, 2 rows like screenshot
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.08),
    ncol=3,
    frameon=False,
    fontsize=16,
    handlelength=1.0,
    handletextpad=0.3,
    columnspacing=0.8,
    markerscale=0.9,
)

plt.tight_layout()
plt.show()
```
